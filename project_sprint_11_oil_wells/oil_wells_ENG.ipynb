# Selection of location for a well

## Description of the project

The project for the production company Glavrosgosneft.It is necessary to decide where to drill a new well.

Oil samples in three regions were provided: in each 10,000 fields, where the quality of oil and the volume of its reserves were measured.We will build a machine learning model that will help determine the region where production will bring the greatest profit.Let us analyze the possible profit and risks of technology *bootstrap. *

Steps for choosing a location:

- in the chosen region, deposits are looking for, for each determine the values ​​of the signs (the quality of oil and the volume of its reserves);
- build a model to predict the volume of stocks in new wells;
- Choose deposits with the highest ratings of values.The number of deposits depends on the budget of the company and the cost of developing one well;
- profit is equal to the total profit of selected deposits.

The characteristics for each well in the region are already known.

Plan Plan:

1. We load and prepare the data.Let us explain the procedure.
2. We will train and check the model for each region:
1. We will divide the data into training and validation samples in the ratio of 75:25.
2. We will train the model and make predictions in the validation sample.
3. Save predictions and correct answers in the validation sample.
4. We print the average stock of predicted raw materials and RMSE model.
5. We analyze the results.
3. We will prepare for the calculation of profit:
1. We will save all key values ​​for calculations in individual variables.
2. We calculate the sufficient volume of raw materials for breaking the development of a new well.Compare the resulting volume of raw materials with the average margin in each region.
3. We draw conclusions at the stage of preparing the calculation of profit.
4. We will write a function for calculating profit from the selected wells and predictions of the model:
1. Choose wells with maximum values ​​of predictions.
2. We make the target value of the volume of raw materials corresponding to these predictions.
3. We calculate the profit for the resulting volume of raw materials.
5. We calculate the risks and profit for each region:
1. We apply the Bootstrap technique with 1000 samples to find a profit distribution.
2. Find average profit, 95%confidential interval and risk of losses.
3. We draw conclusions: we will offer the region for the development of wells and justify the choice.

Data of geological exploration of three regions are in files:
- `/datasets/geo_data_0.csv`
- `/datasets/geo_data_1.csv`
- `/datasets/geo_data_2.csv`

Data description:

- `ID` - a unique well identifier;
- `f0`,` f1`, `f2` - three signs of points (it does not matter what they mean, but the signs themselves are significant);
- `Product` - the volume of stocks in the well (thousand barrels).

Terms of the problem:

- only linear regression is suitable for teaching the model (the rest are insufficiently predictable).
- During the exploration of the region, 500 points are examined, from which, using machine learning, they choose the 200 best for development.
- The budget for the development of wells in the region is 10 billion rubles.
- At the current prices, one barrel of raw materials brings 450 rubles of income.The income from each unit of the product is 450 thousand rubles, since the volume is indicated in thousands of barrels.
- After assessing the risks, only those regions need to be left in which the probability of losses is less than 2.5%.Among them, the region is chosen with the greatest average profit.
- Synthetic data: the details of the contracts and the characteristics of the deposits are not disclosed.

## Learning and checking the model

### Data Sampling

```python
def split(data):
    x = data.drop(['product', 'id'], axis=1)
    y = data['product']
    x_train, x_valid, y_train, y_valid = train_test_split(
        x, y, test_size=0.25, random_state=1337)
    print('Размер тренировочной выборки:')
    print(f'признаки {x_train.shape};')
    print(f'целевой признак {y_train.shape}')
    print(f'это {round((y_train.count() / y.count()) * 100)}% от датасета')
    print()
    print('Размер валидационной выборки:')
    print(f'признаки {x_valid.shape};')
    print(f'целевой признак {y_valid.shape}')
    print(f'это {round((y_valid.count() / y.count()) * 100)}% от датасета')
    return x_train, x_valid, y_train, y_valid
```

```python
x_0_train, x_0_valid, y_0_train, y_0_valid = split(data_0)
```

```python
x_1_train, x_1_valid, y_1_train, y_1_valid = split(data_1)
```

```python
x_2_train, x_2_valid, y_2_train, y_2_valid = split(data_2)
```

### scaling the data

```python
numeric = ['f0', 'f1', 'f2']

scaler = StandardScaler()

scaler.fit(x_0_train[numeric])
x_0_train[numeric] = scaler.transform(x_0_train[numeric])
x_0_valid[numeric] = scaler.transform(x_0_valid[numeric])

scaler.fit(x_1_train[numeric])
x_1_train[numeric] = scaler.transform(x_1_train[numeric])
x_1_valid[numeric] = scaler.transform(x_1_valid[numeric])

scaler.fit(x_2_train[numeric])
x_2_train[numeric] = scaler.transform(x_2_train[numeric])
x_2_valid[numeric] = scaler.transform(x_2_valid[numeric])
```

### We will train models and make a prediction on a validation sample

```python
def train_model(x_train, x_valid, y_train, y_valid):
    model = LinearRegression().fit(x_train, y_train)
    pred_valid = pd.Series(model.predict(x_valid))
    mse = mean_squared_error(y_valid, pred_valid)
    rmse = math.sqrt(mse)
    print('RMSE модели =', round(rmse, 2))
    print('Средний запас предсказанного сырья =', round(pred_valid.mean(), 2), 'тыс. баррелей')
    return model, pred_valid
```

```python
model_0, pred_0_valid = train_model(x_0_train, x_0_valid, y_0_train, y_0_valid)
```

```python
model_1, pred_1_valid = train_model(x_1_train, x_1_valid, y_1_train, y_1_valid)
```

```python
model_2, pred_2_valid = train_model(x_2_train, x_2_valid, y_2_train, y_2_valid)
```

### We will analyze the results

The most accurate model turned out in the second region, RMSE is only 0.88

The third region has the largest number of predicted raw materials

## Preparation for the calculation of profit

```python
PRICE = 450_000
BUDGET = 10_000_000_000

def profit(pred, y):
    state = np.random.RandomState(1337)
    values = []
    for i in range(1000):
        sub = pred.sample(n=500, replace=True, random_state=state)
        best_200_pred = sub.sort_values(ascending=False)[:200]
        best_200 = y.iloc[best_200_pred.index]
        result = PRICE * best_200.sum() - BUDGET
        values.append(result)

    values = pd.Series(values)

    return values

def risk(profit):
    return len(profit[profit<0])/len(profit)*100
```

```python
profit_limit = BUDGET / 200 / PRICE

print('достаточный объём сырья для безубыточной разработки новой скважины', round(profit_limit, 2), 'тыс. баррелей')
print('Средний запас предсказанного сырья первого региона', round(pred_0_valid.mean(), 2), 'тыс. баррелей')
print('Средний запас предсказанного сырья второго региона', round(pred_1_valid.mean(), 2), 'тыс. баррелей')
print('Средний запас предсказанного сырья третьего региона', round(pred_2_valid.mean(), 2), 'тыс. баррелей')
```

Conclusion:

The average stock of raw materials does not reach the required reserve that would bring profit, but this is not critical, because the 200 best wells, not medium, will be selected, how much it will be profitably calculated further

## Calculation of profit and risks

```python
profit_0 = profit(pred_0_valid, y_0_valid)
profit_1 = profit(pred_1_valid, y_1_valid)
profit_2 = profit(pred_2_valid, y_2_valid)

profits = pd.DataFrame({'profit_0': profit_0, 'profit_1': profit_1, 'profit_2': profit_2}) / 1_000_000

conf_0_lo = round(profit_0.quantile(0.025) / 1_000_000, 2)
conf_0_up = round(profit_0.quantile(0.975) / 1_000_000, 2)
conf_1_lo = round(profit_1.quantile(0.025) / 1_000_000, 2)
conf_1_up = round(profit_1.quantile(0.975) / 1_000_000, 2)
conf_2_lo = round(profit_2.quantile(0.025) / 1_000_000, 2)
conf_2_up = round(profit_2.quantile(0.975) / 1_000_000, 2)


print(f'Средняя прибыль первого региона = {round(profit_0.mean()/1_000_000, 2)} млн.руб.')
print(f'95%-ый доверительный интервал первого региона = ({conf_0_lo} - {conf_0_up}) млн.руб.')
print(f'Риск убытков в первом регионе составляет {round(risk(profit_0), 2)}%')
print()
print(f'Средняя прибыль второго региона = {round(profit_1.mean()/1_000_000, 2)} млн.руб.')
print(f'95%-ый доверительный интервал первого региона = ({conf_1_lo} - {conf_1_up}) млн.руб.')
print(f'Риск убытков во втором регионе составляет {round(risk(profit_1), 2)}%')
print()
print(f'Средняя прибыль третьего региона = {round(profit_2.mean()/1_000_000, 2)} млн.руб.')
print(f'95%-ый доверительный интервал первого региона = ({conf_2_lo} - {conf_2_up}) млн.руб.')
print(f'Риск убытков в третьем регионе составляет {round(risk(profit_2), 2)}%')

fig, ax = plt.subplots(figsize=(6,6))
sns.kdeplot(data=profits, fill=True, common_norm=False, alpha=0.4, ax=ax) 
ax.legend(['Регион 1', 'Регион 2', 'Регион 3'])
ax.vlines(0, 0, 0.004, color = 'r', linestyle = '--')
ax.text(25, 0.004, 'Порог прибыльности', color = 'r')
plt.show()
```

```python
# Старый код в приложение к вопросу
confidence_interval_0 = st.t.interval(0.95, len(profits['profit_0'])-1, profits['profit_0'].mean(), profits['profit_0'].sem())
confidence_interval_1 = st.t.interval(0.95, len(profits['profit_1'])-1, profits['profit_1'].mean(), profits['profit_1'].sem())
confidence_interval_2 = st.t.interval(0.95, len(profits['profit_2'])-1, profits['profit_2'].mean(), profits['profit_2'].sem())

print(f'95%-ый доверительный интервал первого региона = ({round(confidence_interval_0[0], 2)} - {round(confidence_interval_0[1], 2)}) млн.руб.')
print(f'95%-ый доверительный интервал второго региона = ({round(confidence_interval_1[0], 2)} - {round(confidence_interval_1[1], 2)}) млн.руб.')
print(f'95%-ый доверительный интервал третьего региона = ({round(confidence_interval_2[0], 2)} - {round(confidence_interval_2[1], 2)}) млн.руб.')
```
